```python
import pandas as pd

data = pd.read_csv('D:/Crop-Recommendation/cotton_yield_synthetic_600.csv')
data.head()
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cat = data.select_dtypes(include = ['bool']).columns
num = data.select_dtypes(include = ['int64','float64']).columns
```

```python
for i in num:
    plt.title(i)
    sns.boxplot(data[i])
    plt.show()
```

```python
for i in cat:
    plt.title(i)
    sns.countplot(x = i , data = data)
    plt.show()
```

```python
for i in num:
    plt.title(i)
    sns.histplot(x = i, data = data , kde = True, bins = 10)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for i in cat:
    data[i] = encoder.fit_transform(data[[i]])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot = True)
plt.show()
```
